```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\Windows Party\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types of columns

df.dtypes
```

```python
#Changing data type for visualisation purposes 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df
```

```python
#Create correct Year column

df['correctyear'] = df['released'].astype(str).str[:4]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Scatter plot of budget vs gross to see correlation

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings (m)')
plt.ylabel('Budget for Film (m)')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x= 'budget', y= 'gross', data=df, scatter_kws= {'color': 'red'}, line_kws= {'color': 'blue'})
```

```python
#Looking at the correlation of all numeric columns

df.corr()
```

```python
#visualisation of correlation matrix with a heatmap

correclation_matrix = df.corr(method= 'pearson')

sns.heatmap(correclation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
#Converting all string values to unique numbers to compare correlation of each movie feature

df_numerised = df

for col in df_numerised:
    if (df_numerised[col].dtype == 'object'):
        df_numerised[col] = df_numerised[col].astype('category')
        df_numerised[col] = df_numerised[col].cat.codes

df_numerised
```

```python
correclation_matrix = df_numerised.corr(method= 'pearson')

sns.heatmap(correclation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')


plt.show()
```

```python
correlation_mat = df_numerised.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
